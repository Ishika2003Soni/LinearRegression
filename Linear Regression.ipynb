-# Mean Normalization

```python
import pandas as pd
```

```python
# Sample wine data (for demonstration)
data = {
    'citric acid': [0.24, 0.26, 0.30, 0.34, 0.28],
    'alcohol': [9.4, 9.8, 10.0, 10.5, 10.2],
    'quality': [5, 5, 6, 7, 6]
}
df = pd.DataFrame(data)
df
```

```python
mean = df["alcohol"].mean()
min_val = df["alcohol"].min()
max_val = df["alcohol"].max()
df["alcohol_normalized"] = (df["alcohol"] - mean) / (max_val - min_val)
df
```

## Mean Normalization

```python
X = [10, 20, 30, 40, 50]
mean_val = sum(X) / len(X)
min_val = min(X)
max_val = max(X)

normalized = [(x - mean_val) / (max_val - min_val) for x in X]
print(normalized)
```

## Robust Scaling

```python
import pandas as pd
from sklearn.preprocessing import RobustScaler

# Load the dataset
df = pd.read_csv(r"C:\Users\chira\Downloads\archive (1)\winequality-red.csv")
df
```

## Example of Frequent Encoding

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# Sample data
data = {
    'Customer_ID': [1, 2, 3, 4, 5],
    'Gender': ['Male', 'Female', 'Nan', 'Male', 'Nan']
}

df = pd.DataFrame(data)
df
```

```python
# Convert string 'Nan' to actual np.nan
df['Gender'] = df['Gender'].replace('Nan', np.nan)
df
```

```python
# Step 1: Create SimpleImputer object with 'most_frequent' strategy
imputer = SimpleImputer(strategy='most_frequent')
df
```

```python
df[['Gender']] = imputer.fit_transform(df[['Gender']])
print(df)
```

## Example of Target Encoding

```python
import pandas as pd

# Step 1: Data banaate hain
data = {
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Bangalore', 'Mumbai', 'Bangalore'],
    'House_Price': [50, 70, 60, 40, 80, 50]
}
df = pd.DataFrame(data)
df
```

```python
# Step 3: Har city ko uske average se replace karo
df['City_Encoded'] = df['City'].map(mean_encoded)
df
```

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

# Step 1: Sample data
df = pd.DataFrame({
    'Education': ['High School', 'Bachelor', 'Master', 'PhD', 'Bachelor']
})

# Step 2: Define order manually (important step)
education_order = [['High School', 'Bachelor', 'Master', 'PhD']]

# Step 3: Create encoder object
encoder = OrdinalEncoder(categories=education_order)

# Step 4: Apply encoding
df['Education_Encoded'] = encoder.fit_transform(df[['Education']])

# Output
print(df)
```
